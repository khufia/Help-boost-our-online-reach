We converted the parsed text to a csv file, in order to avoid preprocessing again and again.


```python
from bs4 import BeautifulSoup
for i in range(len(htmlfiles)):
  path = "gdrive/MyDrive/ml-term-project/term-project/help-boost-our-online-reach/html_content/" + htmlfiles[i]
  fr = codecs.open(path, 'r', encoding='utf-8', errors='ignore')
  html_content = fr.read()
  print((i*100)/7395, "%")
  soup = BeautifulSoup(html_content, 'html.parser')
  # kill all script and style elements
  for script in soup(["script", "style"]):
      script.extract()    # rip it out

  # get text
  text = soup.get_text()

  # break into lines and remove leading and trailing space on each
  lines = (line.strip() for line in text.splitlines())
  # break multi-headlines into a line each
  chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
  # drop blank lines
  text = ' '.join(chunk for chunk in chunks if chunk)
  html_data.append({'id': htmlfiles[i], 'text': text})
```

Text Preprocessing.

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
import string
string.punctuation

def remove_punct(text):
  text1 = "".join([char for char in text if char not in string.punctuation])
  return text1

html_content_df["text_clean"] = html_content_df["text"].apply(lambda x: remove_punct(x))
```

Tokenization

```python
import re

def tokenize(text):
  tokens = re.sub('[^a-zA-Z]', ' ', text)
  tokens = re.split('\W+', tokens)
  tokens = [string for string in tokens if string != '']
  return tokens

html_content_df['text_tokenized'] = html_content_df['text_clean'].apply(lambda x: tokenize(x.lower()))
```

```python
html_content_df
```

Removing Stopwords

```python
import nltk

stopword = nltk.corpus.stopwords.words('english')

def remove_stopwords(text):
  text = [word for word in text if word not in stopword]
  return text

html_content_df['text_nostop'] = html_content_df['text_tokenized'].apply(lambda x: remove_stopwords(x))
```

Stemming, instead of Lemmatization.

```python
ps = nltk.PorterStemmer()

def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text

html_content_df['text_stemmed'] = html_content_df['text_nostop'].apply(lambda x: stemming(x))
```

```python
html_content_df['text_stemmed_string'] = html_content_df['text_stemmed'].apply(lambda x: ' '.join(x))
```

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

```python
X_train_vect_avg = []
for v in X_train_vect:
    if v.size:
        X_train_vect_avg.append(v.mean(axis=0))
    else:
        X_train_vect_avg.append(np.zeros(100, dtype=float))
        
X_test_vect_avg = []
for v in X_test_vect:
    if v.size:
        X_test_vect_avg.append(v.mean(axis=0))
    else:
        X_test_vect_avg.append(np.zeros(100, dtype=float))
```

Using Word2Vec With RandomForestClassifier 

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf_model = rf.fit(X_train_vect_avg, y_train.values.ravel())
```

```python
y_pred = rf_model.predict(X_test_vect_avg)
```

```python
from sklearn.metrics import precision_score, recall_score
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print('Precision: {} / Recall: {} / Accuracy: {}'.format(round(precision, 3), round(recall, 3), round((y_pred==y_test).sum()/len(y_pred), 3)))
```
