removing punctuation from the text

```python
import string
string.punctuation

def remove_punct(text):
  text1 = "".join([char for char in text if char not in string.punctuation])
  return text1

df["body_clean"] = df["body"].apply(lambda x: remove_punct(x))
df["title_clean"] = df['title'].apply(lambda x: remove_punct(x))

df
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

tokenizing the text

```python
import re

def tokenize(text):
  tokens = re.sub('[^a-zA-Z]', ' ', text)
  tokens = re.split('\W+', tokens)
  tokens = [string for string in tokens if string != '']
  return tokens

df['body_tokenized'] = df['body_clean'].apply(lambda x: tokenize(x.lower()))
df['title_tokenized'] = df['title_clean'].apply(lambda x: tokenize(x.lower()))
df['webpageDescription'] = df['webpageDescription'].apply(lambda x: tokenize(x.lower()))
```

removing stopwords

```python
import nltk

stopword = nltk.corpus.stopwords.words('english')

def remove_stopwords(text):
  text = [word for word in text if word not in stopword]
  return text

df['body_nostop'] = df['body_tokenized'].apply(lambda x: remove_stopwords(x))
df['title_nostop'] = df['title_tokenized'].apply(lambda x: remove_stopwords(x))
df['webpageDescription'] = df['webpageDescription'].apply(lambda x: remove_stopwords(x))
```

Stemming

```python
ps = nltk.PorterStemmer()

def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text

df['body_stemmed'] = df['body_nostop'].apply(lambda x: stemming(x))
df['title_stemmed'] = df['title_nostop'].apply(lambda x: stemming(x))
df['webpageDescription'] = df['webpageDescription'].apply(lambda x: stemming(x))

df.head()
```

Lemmatizing

```python
import re
wn = nltk.WordNetLemmatizer()

def lemmatizing(text):
  text1 = [wn.lemmatize(word) for word in text]
  return text1

df['body_lemmatized'] = df['body_nostop'].apply(lambda x: lemmatizing(x))
df['title_lemmatized'] = df['title_nostop'].apply(lambda x: lemmatizing(x))
# df['webpageDescription'] = df['webpageDescription'].apply(lambda x: lemmatizing(x))
```
